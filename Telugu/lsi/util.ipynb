```python
vowels=set("అ ఆ ఇ ఈ ఉ ఊ ఋ ౠ ఌ ౡ ఎ ఏ ఐ ఒ ఓ ఔ అం అః".split())

conso="క	ఖ	గ	ఘ	ఙ	చ	ఛ	జ	ఝ	ఞ	ట	ఠ	డ	ఢ	ణ	త	థ	ద	ధ	న	ప	ఫ	బ	భ	మ	య	ర	ల	వ	శ	ష	స	హ	ళ	క్ష	ఱ"

matra = "ఀ	ఁ	ం	ః	ఄ	ఽ	ా	ి	ీ	ు	ూ	ృ	ౄ	ె	ే	ై	ొ	ో	ౌ	్	ౕ	ౖ	ౢ	ౣ"

halantha='్'
```

```python
consonants=set(conso.split("\t"))
matra = set(matra.split("\t"))

vowels = list(vowels)
consonants=list(consonants)
matra=list(matra)

alphabets = vowels+consonants
```

```python
print(alphabets)
```

```python
def remove_special_characters(text):
    string=''
    for i in text:
        if i in alphabets+matra:
            string=string+i
        else:
            string=string+" "
    return " ".join(string.split())
```

```python
lsi = models.LsiModel.load('lsi.model')
```

```python
dictionary=corpora.Dictionary.load('dictionary')
corpus=corpora.MmCorpus('corpus')
```

```python
data=pd.read_csv('../telugu.csv')
```

```python
def search(query,counter=10):
    query=remove_special_characters(query)
    words=query.split()
    words = [word for word in words if word not in stopwords]
    vec_bow = dictionary.doc2bow(words)
    vec_lsi = lsi[vec_bow]
    index = similarities.MatrixSimilarity(lsi[corpus],num_features=30)
    scoring = index[vec_lsi]
    verse_index=np.argsort(scoring)[-1::-1][:counter]
    return verse_index
```

```python
query='భగవద్గీత ప్రాముఖ్యత'
```

```python
verse_index=search(query)
data.loc[verse_index]
```
